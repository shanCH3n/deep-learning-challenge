## Step 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("Resources/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(['EIN', 'NAME'], axis=1)
```

```python
# Determine the number of unique values in each column.
print(application_df.nunique())
```

```python
# Look at APPLICATION_TYPE value counts for binning
print(application_df['APPLICATION_TYPE'].value_counts())
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
cutoff = 500
application_counts = application_df['APPLICATION_TYPE'].value_counts()
application_types_to_replace = list(application_counts[application_counts < cutoff].index)

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
clsf_counts = application_df["CLASSIFICATION"].value_counts()
clsf_counts
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
class_counts_oneUP = clsf_counts[clsf_counts > 1]
class_counts_oneUP
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
cutoff = 1000
classifications_to_replace = list(clsf_counts[clsf_counts < cutoff].index)

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Check datatypes
print(application_df.dtypes)
```

```python
# Examine categorical variables
print(application_df['USE_CASE'].value_counts())
print('-------------------------------------------')
print(application_df['STATUS'].value_counts())
print('-------------------------------------------')
print(application_df['INCOME_AMT'].value_counts())
print('-------------------------------------------')
print(application_df['IS_SUCCESSFUL'].value_counts())
print('-------------------------------------------')
```

#### Likely data imbalance for STATUS variable. KIV for Model Optimisation.

```python
# Convert categorical data to numeric with `pd.get_dummies`
# Using one-hot encoding
df_OneHot = pd.get_dummies(application_df)
df_OneHot
```

```python
# Split our preprocessed data into our features and target arrays
# The target variable is 'IS_SUCCESSFUL'.
y = df_OneHot['IS_SUCCESSFUL'].values
X = df_OneHot.drop(['IS_SUCCESSFUL'], axis=1).values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Step 2: Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=7, input_dim=len(X_train.columns), activation="tanh"))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=14, input_dim=len(X_train.columns), activation='relu'))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Define a callback function that saves the model's weight every five epochs.
model1_weights = tf.keras.callbacks.ModelCheckpoint(
    filepath='./model01/weights',
    verbose= 1,
    save_weights_only= True,
    mode='auto',
    save_freq='epoch',
    period=5
)
```

```python
# Train the model
batch_size = 32
epochs = 100
nn.fit(X_train_scaled, y_train, 
       batch_size=batch_size, 
       epochs=epochs, callbacks=[model1_weights])
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
